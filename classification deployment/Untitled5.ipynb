extracting number of days from date of admission to discharge date

```python
df['Date of Admission']= pd.to_datetime(df['Date of Admission'])
df['Discharge Date']= pd.to_datetime(df['Discharge Date'])
df['Days hospitalized'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
df.head()
```

```python
# KNN
x=df.drop(['Room Number','Billing Amount','Days hospitalized','Test Results'],axis=1)
y=df['Test Results']
```

```python
x
```

```python
from sklearn import preprocessing
x=preprocessing.StandardScaler().fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33,random_state=42)
```

```python
ks=25
mean_acc=np.zeros((ks-1))
# train as well as predict
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
for n in range(1,ks-1):
    neigh=KNeighborsClassifier(n_neighbors=n).fit(x_train,y_train)
    yhat=neigh.predict(x_test)
    mean_acc[n-1]=accuracy_score(y_test,yhat)
```

```python
mean_acc
```

```python
print("best accuracy was with",mean_acc.max(),"with k=",mean_acc.argmax())
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnmodel=KNeighborsClassifier(n_neighbors=21)
knnmodel.fit(x_train,y_train)
y_predict1=knnmodel.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_test,y_predict1)
acc
```

```python
# logistic
x=df.drop(['Room Number','Billing Amount','Days hospitalized','Test Results'],axis=1)
y=df['Test Results']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[0.5, 0.6, 0.7, 0.8],'max_iter':[100,200,300]}
```

```python
classifier_regressor=GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
y_predi=classifier_regressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_predi,y_test)
score
```

```python
# classification_report(y_predi,y_test)
print(classification_report(y_test,y_predi))
```

```python
# SVM
x=df.drop(['Room Number','Days hospitalized','Billing Amount','Test Results','Insurance Provider'],axis=1)
y=df['Test Results']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVC
model=SVC(kernel="linear")
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
accuracy_score(y_test,y_pred)
```
