```python
X=df.drop('quality',axis=1)
y=df['quality']
```

```python
# RANDOM FOREST
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state= 42)
```

```python
rfc=RandomForestRegressor()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_rfc_pred=rfc.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_rfc_pred)
r2
```

```python
# LINEAR
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
mse
```

```python
reg_pred=regression.predict(X_test)
reg_pred
```

```python
from sklearn import metrics
r2_score=metrics.r2_score(y_test,reg_pred)
print("R2 score=",r2_score)
```

```python
# SVM
from sklearn.svm import SVR
model=SVR(kernel="linear")
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred=model.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# KNN
from sklearn import preprocessing
X=preprocessing.StandardScaler().fit_transform(X)
X
```

```python
ks=25
mean_acc=np.zeros((ks-1))
# train as well as predict
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
for n in range(1,ks-1):
    neigh=KNeighborsRegressor(n_neighbors=n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1]=r2_score(y_test,yhat)
```

```python
mean_acc
```

```python
print("best accuracy was with",mean_acc.max(),"with k=",mean_acc.argmax())
```

```python
from sklearn.neighbors import KNeighborsRegressor
knnmodel=KNeighborsRegressor(n_neighbors=20)
knnmodel.fit(X_train,y_train)
y_predict1=knnmodel.predict(X_test)
```

```python
r2=r2_score(y_test,y_predict1)
r2
```

THEREFORE MAXIMUM ACCURACY IS 42% WITH RANDOM FOREST
